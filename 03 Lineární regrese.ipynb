# Lineární (a polynomiální) regrese

inspirace: https://github.com/martinpilat/CImethods

```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
import numpy as np

def fce(x):
    return 2*x+3

def rand_data(f, noise=2):
    x = np.random.random(100)*10-5
    y = f(x) + np.random.randn(100)*noise
    return x, y

def plot_data(x, y, f=None):
    plt.figure(figsize=(12,8))
    plt.scatter(x, y)
    if f:
        ln_x = np.linspace(-6, 6)
        ln_y = f(ln_x)
        plt.plot(ln_x, ln_y)

x, y = rand_data(fce, 4)
plot_data(x, y, fce)
plt.show()
```

```python
def fce(x):
    return 2*x+3

x, y = rand_data(fce, 1)
plot_data(x, y)
```

```python
p = np.polyfit(x, y, 1)
l = np.poly1d(p)
plot_data(x, y, l)
```
