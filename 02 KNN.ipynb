```python
%matplotlib inline
import csv
import matplotlib.pyplot as plt

data = list(csv.reader(open('data/iris.data')))

setosa = data[0:50]
versicolor = data[50:100]
virginica = data[100:150]

def plot_irises(data, x_attr, y_attr, color):
    xs = [a[x_attr] for a in data]
    ys = [a[y_attr] for a in data]
    plt.plot(xs, ys, 'o', color=color)

plot_irises(setosa, 0, 1, 'r')
plot_irises(versicolor, 0, 1, 'g')
plot_irises(virginica, 0, 1, 'b')
```
